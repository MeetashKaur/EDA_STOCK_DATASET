```python
import pandas as pd
import numpy as np
import plotly as py
from plotly.offline import iplot
import cufflinks as cf
import plotly.tools as tls
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("C:/Users/DELL/Technocolabs(DS Intern)/HistoricalData_FB.csv",index_col="Date")
data.head()
```

```python
data.tail()
```

```python
reversed_data=data.iloc[::-1]
reversed_data.head()
```

```python
reversed_data.shape
```

```python
reversed_data.dtypes
```

```python
reversed_data.isnull().sum()
```

```python
reversed_data['N_Open']=reversed_data["Open"].str.replace("$","").astype("float")
reversed_data['N_Low']=reversed_data["Low"].str.replace("$","").astype("float")
reversed_data['N_High']=reversed_data["High"].str.replace("$","").astype("float")
reversed_data['N_Close']=reversed_data["Close/Last"].str.replace("$","").astype("float")
```

```python
reversed_data.head()
```

```python
col=["Close/Last","Open","High","Low"]
for i in col:
    reversed_data=reversed_data.drop([i],axis=1)
```

```python
reversed_data.head()
```

```python
reversed_data.dtypes
```

# EDA

#Univariate Analysis

```python
reversed_data.iplot(y="Volume",yTitle='Volume',color='red',title='History of Volume')
```
